```python
##Feature Engineering (label encoding and onehot encoding)
```

```python
df.columns
```

```python
#independent and dependent features
X=df[['tip','sex','smoker','day','time','size']] #independent
Y=df['total_bill'] #dependent
```

```python
##train-test-split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
X_train
```

```python
Y_train
```

```python
for col in ['sex', 'smoker', 'time']:
    X_train[col] = X_train[col].str.strip().str.lower()
    X_test[col]  = X_test[col].str.strip().str.lower()
```

```python
#feature encoding
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le3 = LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
#here we dont fit as it is a test data
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
```

```python
#Onehot Encoding for Date -- columnTransformer

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('onehot',OneHotEncoder(drop='first'),[3])],remainder='passthrough')
```

```python
ct.fit_transform(X_train)
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
##SVM -- Support Vector Regression
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,Y_train)
```

```python
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
print(r2_score(Y_test,y_pred))
print(mean_absolute_error(Y_test,y_pred))
```
